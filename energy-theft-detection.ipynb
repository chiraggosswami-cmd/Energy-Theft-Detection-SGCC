# Let's split the dataset first

```python
# Remove the ID column
df.drop(columns="CONS_NO", axis=1,inplace=True)

# Separate features and target
X = df.drop("FLAG", axis=1)
y = df["FLAG"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
from sklearn.impute import SimpleImputer
from imblearn.over_sampling import SMOTE

# 1. Initialize the imputer to use the median
imputer = SimpleImputer(strategy="median")

# 2. Fit on X_train and transform both train and test sets to avoid data leakage
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

# 3. Now apply SMOTE on the imputed training data
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train_imputed, y_train)
```

```python
y_train_smote.value_counts()
```

No more class imbalance

# Checking Decision Tree and Random Forest

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# 1. Initialize the models
dt_model = DecisionTreeClassifier(random_state=42)
rf_model = RandomForestClassifier(random_state=42, n_estimators=100)

# 2. Train on the SMOTE-resampled (but unscaled) training data
dt_model.fit(X_train_smote, y_train_smote)
rf_model.fit(X_train_smote, y_train_smote)

# 3. Predict on the test set
y_pred_dt = dt_model.predict(X_test_imputed)
y_pred_rf = rf_model.predict(X_test_imputed)

# 4. Evaluate
print("=== Decision Tree Performance ===")
print(classification_report(y_test, y_pred_dt))

print("\n=== Random Forest Performance ===")
print(classification_report(y_test, y_pred_rf))
```

# Feature Engineering

```python
# Creating a copy to avoid warnings
df_feat = df.copy()

# Drop ID and keep target separate
df_consumption = df_feat.drop(columns=["CONS_NO", "FLAG"])

# 1. Total & Average Consumption per consumer
df_feat['total_consumption'] = df_consumption.sum(axis=1)
df_feat['mean_consumption'] = df_consumption.mean(axis=1)
df_feat['median_consumption'] = df_consumption.median(axis=1)

# 2. Volatility (Thiefs often have erratic or sudden drops in usage)
df_feat['std_consumption'] = df_consumption.std(axis=1)

# 3. Drop in consumption (Comparing max usage to min usage)
df_feat['max_consumption'] = df_consumption.max(axis=1)
df_feat['min_consumption'] = df_consumption.min(axis=1)
df_feat['consumption_range'] = df_feat['max_consumption'] - df_feat['min_consumption']

# 4. Zero consumption days (Many thieves bypass the meter entirely, dropping usage to 0)
df_feat['zero_days_count'] = (df_consumption == 0).sum(axis=1)

# Now extract your new X and y
features = ['total_consumption', 'mean_consumption', 'median_consumption', 
            'std_consumption', 'consumption_range', 'zero_days_count']

X = df_feat[features]
y = df_feat['FLAG']
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train_smote)

X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

# 1. Initialize the models (using a fast linear/RBF kernel for SVM)
knn_model = KNeighborsClassifier(n_neighbors=5)
svm_model = SVC(kernel='rbf', random_state=42) 

# 2. Train on the SCALED training data
knn_model.fit(X_train_scaled, y_train_smote)
svm_model.fit(X_train_scaled, y_train_smote)

# 3. Predict on the SCALED test set
y_pred_knn = knn_model.predict(X_test_scaled)
y_pred_svm = svm_model.predict(X_test_scaled)

# 4. Evaluate
print("=== KNN Performance (Scaled) ===")
print(classification_report(y_test, y_pred_knn))

print("\n=== SVM Performance (Scaled) ===")
print(classification_report(y_test, y_pred_svm))
```
